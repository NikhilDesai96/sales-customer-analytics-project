```python
#Create sales segments(High Sales,Medium Sales,Low Sales):

query="""
Select Sales,

Case
  When Sales>1000 Then 'High Sales'
  When Sales>500 Then 'Medium Sales'
  Else 'Low Sales'
END as Sales_Category

from Sales_data
limit 10;
"""
pd.read_sql(query,conn)

```

```python
#Profit Status
query="""
select Profit,

Case
  when Profit>0 Then 'Profit'
  else 'Loss'
End As Profit_Status

from sales_data
limit 10;
"""
pd.read_sql(query,conn)
```

## Real Business SQL Analysis

```python
#Top 10 Customers by Sales

query="""
Select "Customer Name",
       Sum(Sales)As Total_Sales
From sales_data
Group By "Customer Name"
Order By Total_Sales Desc
limit 10;
"""
pd.read_sql(query,conn)
```

Purpose
- To identify top revenue-generating customers.
- To understand customer contribution to total sales.
- To support customer retention strategies.

```python
#Most Profitable Categories
query="""
select Category,sum(Profit)As Total_Profit
from sales_data
group by Category
order by Total_Profit Desc;
"""
pd.read_sql(query,conn)
```

Purpose
- To identify categories generating maximum profit.
- To understand overall category performance.
- To support strategic business decisions.

```python
#Loss-Making Sub-Categories
query="""
select "Sub-Category",
      Sum(Profit)as Total_Profit
from sales_data
Group by "Sub-Category"
Having Sum(Profit)<0
Order By Total_Profit;
"""
pd.read_sql(query,conn)
```

Purpose
- To identify products causing business losses.
- To detect weak-performing product segments.
- To support inventory and pricing decisions.

```python
#Monthly Sales Trend
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Order Month"]=df["Order Date"].dt.month

df.to_sql("sales_data",conn,if_exists="replace",index=False)
```

```python


query="""
select
  strftime('%m',"Order Date")As Order_Month,
  Sum(Sales)as Total_Sales
from sales_data
group by Order_Month
order by Order_Month;
"""
pd.read_sql(query,conn)
```

Purpose
- To analyze monthly sales performance.
- To identify seasonal sales trends.
- To support forecasting and business planning.

```python
#Discount Impact on Profit
query="""
select Discount, Avg(Profit)as Average_Profit
from sales_data
group by Discount
order by Discount;
"""
pd.read_sql(query,conn)
```

Purpose
- To analyze relationship between discount and profitability.
- To understand whether high discounts hurt business profit.
- To support pricing strategy decisions.

```python
#Shipping Performance Analysis
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df["Shipping_Days"]=(df["Ship Date"]-df["Order Date"]).dt.days
df.to_sql("sales_data",conn,if_exists="replace",index=False)
```

```python
query="""
select Region,
   Avg(Shipping_Days) as Avg_Shipping_Days
from sales_data
group by Region
order by Avg_Shipping_Days Desc;
"""
pd.read_sql(query,conn)
```

Purpose
- To compare delivery performance across regions.
- To identify regions with slower shipping efficiency.
- To support logistics optimization.


```python
#Customer Segmentation
query="""
select "Customer Name",
      Sum(Sales)as Total_Sales,
CASE
  When Sum(Sales)>10000 Then 'High Value'
  When Sum(Sales)>5000 Then 'Medium Value'
  Else 'Low Value'
End as Customer_Segment
from sales_data
group by "Customer Name"
order by "Total Sales" Desc;
"""
pd.read_sql(query,conn)
      
```

Purpose
- To segment customers based on spending behavior.
- To identify high-value customers.
- To support targeted marketing and retention strategies.
