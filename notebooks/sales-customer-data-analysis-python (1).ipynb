## 8. Visualization EDA

```python
# Sales Distribution Analysis
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"],bins=30, kde= True)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### Purpose

- To understand the distribution of sales values across all orders.
- To identify whether most transactions are low-value or high-value.
- To detect skewness and potential outliers in sales data.
- To analyze customer purchasing patterns.

```python
# Category-wise Sales Comparison
plt.figure(figsize=(8,5))
sns.barplot(x="Category",y="Sales", data=df)
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

```

### Purpose
- To compare total sales performance across different product categories.
- To identify which category contributes the highest sales.
- To understand category-level business performance.
- To support decision-making for inventory and marketing focus.

```python
## Sales Contribution by Category
df.groupby("Category")["Sales"].sum().plot(kind="pie",autopct="%1.1f%%",figsize=(6,6))
plt.title("Sales Contribution by Category")
plt.ylabel("")
plt.show()
```

### Purpose 

- To analyze the percentage contribution of each category to overall sales.
- To understand which category dominates total business revenue.
- To visualize category share in an easy-to-understand format.
- To identify major and minor contributors to company sales.

```python
## Sales vs Profit Relationship
plt.figure(figsize=(8,5))
plt.scatter(df["Sales"],df["Profit"])
plt.title("Sales vs Profit Relationship")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

### Purpose 

- This graph is used to understand the relationship between Sales and Profit.
- It helps identify whether higher sales generate higher profit and also shows loss-making transactions or outliers.


```python
## Monthly Sales Trend

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"]=df["Order Date"].dt.month
monthly_sales = df.groupby("Month")["Sales"].sum()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### Purpose
- To visualize monthly sales trends over time. 
- To identify peak and low sales months. 
- To analyze seasonality in business performance. 
- To support forecasting and strategic planning

```python
## Correlation Analysis

corr_matrix=df[["Sales","Profit","Discount","Quantity"]].corr()
plt.figure(figsize=(8,5))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Purpose
- To visualize correlations between important business variables.
- To identify positive and negative relationships.
- To understand factors affecting sales and profit.
- To support data-driven business insights.

## 9. Feature Engineering

```python
df["Profit_Margin"]=(df["Profit"]/df["Sales"])
```

```python
df.head()
```

```python
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df["Shipping_Days"]=(df["Ship Date"]- df["Order Date"]).dt.days
```

```python
df.head()
```

## 10. Advanced Business Insights

```python
## Top 10 Customers by Sales
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending= False).head(10).reset_index()
top_customers
```

```python
## Most Profitable Categories
profit_by_category= df.groupby("Category")["Profit"].sum().sort_values(ascending= False).reset_index()


profit_by_category
```

```python
## Loss-Making Sub-Categories

loss_making_by_category=df.groupby("Sub-Category")["Profit"].sum().sort_values()

loss_making_by_category= loss_making_by_category[loss_making_by_category<0].sort_values().reset_index()

loss_making_by_category
```

```python
## Average Shipping Days by Region

df.groupby("Region")["Shipping_Days"].mean().sort_values().reset_index()
```

```python
##Discount Impact on Profit

df.groupby("Discount")["Profit"].mean().reset_index()
```

## 11. Final Business Conclusions

- Technology category generated the highest sales and profit.
- Certain sub-categories such as Tables and Bookcases contributed to overall losses.
- Higher discounts negatively impacted profitability.
- Sales performance showed seasonal fluctuations across months.
- A small group of customers generated significant revenue contribution.
- Shipping performance varied across regions, indicating operational differences.
